##Temp View - want to use sql on top of dataframe, within session in one notebook

```python
df_tempView=spark.sql("select * from databricksshashi.silver.product_enr")
```

```python
df_tempView.createOrReplaceTempView("temp_product")
```

```python
display(spark.sql("select * from temp_product limit 5"))
```

```python
display(spark.sql("""
         select *,
          CASE
            when category ='Toys' THEN 'YES' ELSE 'NO' END as flag
            from temp_product limit 5
          """))
```
